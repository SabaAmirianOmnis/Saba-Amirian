```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd


hours = np.array ( [ 1.0 , 2.0 , 2.5 , 3.0 , 3.5 , 4.0 , 1.5 ] ).reshape( -1 , 1 )


scores = np.array ( [ 70 , 80 , 85 , 90 , 88 , 95 , 75 ] )


model = LinearRegression ()
model.fit ( hours , scores )


predicted_scores = model . predict ( hours )


plt.scatter ( hours , scores , color = 'blue' , label = 'Actual Scores' )
plt.plot ( hours , predicted_scores , color = 'red' , label = 'Regression Line' )
plt.title ( "Study Hours VS Score - Saba 2025" )
plt.xlabel ( "Hours Studied" )
plt.ylabel ( "Score" )
plt.legend ()
plt.show ()


accuracy = model.score ( hours , scores )
print ( f"Model accuracy : { accuracy:.2f}" )
```
