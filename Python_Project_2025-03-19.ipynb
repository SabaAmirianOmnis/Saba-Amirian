```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler



df = pd.read_csv(r"C:\Users\ASUS\Downloads\Iris.csv")



scaler = StandardScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]), columns = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalLengthCm"])



plt.boxplot(df_normalized["SepalLengthCm"])
plt.title("Box Plot of Normalized Sepal Length")
plt.ylabel("Normalized Value")
plt.show()
```
