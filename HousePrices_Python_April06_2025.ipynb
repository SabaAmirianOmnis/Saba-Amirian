```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns




df = pd.read_csv(r"C:\Users\ASUS\Downloads\House Prices.csv")




columns = ["LotArea", "OverallQual", "YearBuilt", "SalePrice"]
df = df[columns]




stats_list = [("Average SalePrice", df["SalePrice"].mean()), ("Average LotArea", df["LotArea"].mean()), ("Average OverallQual", df["OverallQual"].mean())]




print("Statistics List :")
for stat_name, stat_value in stats_list :
    print(f"{stat_name} : {stat_value}")




plt.figure(figsize = (8, 5))
sns.boxplot(x = df["SalePrice"], color = "lightblue")
plt.title("Box Plot of SalePrice")
plt.xlabel("SalePrice ($)")
plt.savefig("HousePrices_BoxPlot_April06_2025.png")
plt.show()




plt.figure(figsize = (8, 5))
sns.boxplot(x = "OverallQual", y = "SalePrice", data = df, palette = "Set2")
plt.title("SalePrice by OverallQual")
plt.savefig("HousePrices_BoxPlotByQual_April06_2025.png")
plt.show()
```
