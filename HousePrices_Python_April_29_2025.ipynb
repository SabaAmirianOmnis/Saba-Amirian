```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\House Prices.csv")
```

```python
df['Date'] = pd.to_datetime(df['YrSold'].astype(str) + '-' + df['MoSold'].astype(str) + '-01')
```

```python
price_over_time = df.groupby('Date')['SalePrice'].mean().reset_index()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(price_over_time['Date'], price_over_time['SalePrice'], marker = 'o')
plt.title('Average Sale Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Sale Price')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
