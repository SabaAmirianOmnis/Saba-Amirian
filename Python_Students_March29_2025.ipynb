```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\StudentsPerformance.csv")




def average_score(column) :
    return df.groupby(column)['math score'].mean()




avg_by_gender = average_score = average_score('gender')
print("Average Math Score by Gender :\n", avg_by_gender)




sns.barplot(x = 'gender', y = 'math score', data = df)
plt.title('Average Math Score by Gender')
plt.show()
```
