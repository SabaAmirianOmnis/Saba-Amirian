```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt



df = pd.read_csv(r"C:\Users\ASUS\Documents\Study_Data.csv")



X = df[["Hours Studied"]]
Y = df["Pass (1=Yes, 0=No)"]



model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,Y)



predictions = model.predict(X)



accuracy = accuracy_score(Y,predictions)
print(f"Model accuracy:{accuracy*100:.2f}%")



plt.scatter(X,Y,color="blue",label="Actual Data")
plt.scatter(X,predictions,color="red",marker="x",label="Predictions")
plt.title("Hours Studied vs Pass/Fail (KNN)")
plt.xlabel("Hours Studied")
plt.ylabel("Pass (1=Yes, 0=No)")
plt.legend()
plt.show()
```
