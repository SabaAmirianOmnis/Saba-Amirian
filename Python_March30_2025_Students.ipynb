```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\StudentsPerformance.csv")




scores_dict = df.groupby('gender')['math score'].mean().to_dict()
print("Average Math Scores :", scores_dict)




plt.bar(scores_dict.keys(), scores_dict.values(), color = ['pink', 'bl'])
plt.title('Average Math Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.show()
```
