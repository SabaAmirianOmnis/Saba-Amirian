```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\House Prices.csv")




print(df[["SalePrice", "GrLivArea", "OverallQual"]].head())
```

```python
plt.scatter(df["GrLivArea"], df["SalePrice"], c = df["OverallQual"], cmap = "coolwarm")
plt.title("House Price v Living Area")
plt.xlabel("Living Area (sq ft)")
plt.ylabel("Sale Price ($)")
plt.colorbar(label = "Overall Quality")
plt.savefig("Scatter_April_12_2025.png")
plt.show()
```
