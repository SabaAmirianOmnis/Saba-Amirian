```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\Medical.csv")
def average_charges(column) :
    return df.groupby(column)['charges'].mean()




avg_by_smoker = average_charges('smoker')
print("Average Charges by Smoker :\n", avg_by_smoker)




sns.barplot(x = 'smoker', y = 'charges', data = df)
plt.title('Average Charges by Smoker')
plt.show()
```
