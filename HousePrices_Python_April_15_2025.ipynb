```python
import pandas as pd




df = pd.read_csv(r"C:\Users\ASUS\Downloads\House Prices.csv")




print(df[['OverallQual', 'GrLivArea', 'SalePrice']].head())




def count_high_price(data, threshold, index = 0):
    if index >= len(data):
        return 0
    count = 1 if data['SalePrice'].iloc[index] > threshold else 0
    return count + count_high_price(data, threshold, index + 1)




threshold = 200000
result = count_high_price(df, threshold)
print(f"Number of houses with price > $ {threshold} : {result}")
df.describe().style.set_caption




import seaborn as sns
import matplotlib.pyplot as plt




sns.scatterplot(x = 'OverallQual', y = 'SalePrice', data = df)
plt.title('House Price vs Quality')
plt.show()
```
