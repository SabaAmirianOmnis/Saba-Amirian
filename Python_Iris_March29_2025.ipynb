```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\Iris.csv")
def average_petal(column) :
    return df.groupby(column)['PetalLengthCm'].mean()




avg_by_Id = average_petal('Id')
print("Average Petal Length by Id :\n", avg_by_Id)




sns.scatterplot(x = 'PetalWidthCm', y = 'PetalLengthCm', hue = 'Id', data = df)
plt.title('Petal Width V Length by Id')
plt.show()
```
