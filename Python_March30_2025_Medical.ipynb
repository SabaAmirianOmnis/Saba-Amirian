```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\Medical.csv")




charges_dict = df.groupby('smoker')['charges'].mean().to_dict()
print("Average Charges :", charges_dict)




plt.bar(charges_dict.keys(), charges_dict.values(), color = ['green', 'red'])
plt.title('Average Charges by Smoker Status')
plt.xlabel('Smoker')
plt.ylabel('Charges')
plt.show()
```
