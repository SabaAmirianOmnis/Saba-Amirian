```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt



data = {'Hours Studied':[1.0,2.0,2.5,3.0,3.5,4.0,1.5],'Practice Tests':[1,2,2,3,3,4,1],'Score':[70,80,85,90,88,95,75]}



df = pd.DataFrame(data)



X = df[['Hours Studied','Practice Tests']]
Y = df['Score']



model = LinearRegression()
model.fit(X,Y)



predicted_scores = model.predict(X)



print("Coefficients:",model.coef_)
print("Intercept:",model.intercept_)



accuracy = model.score(X,Y)
print(f"Model accuracy:{accuracy:.2f}")



plt.scatter(Y,predicted_scores,color = 'blue')
plt.plot([Y.min(),Y.max()],[Y.min(),Y.max()],color = 'red',linestyle='--')
plt.title("Actual vs Predicted Scores - Saba 2025")
plt.xlabel("Actual Score")
plt.ylabel("Predicted Score")
plt.show()
```
