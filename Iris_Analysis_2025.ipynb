```python
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv(r"C:\Users\ASUS\Downloads\Iris.csv")


print(df.head())



print(df.describe())



print(df.isnull().sum())
```

```python
plt.scatter(df["SepalLengthCm"], df["SepalWidthCm"], c = df["Species"].astype('category').cat.codes, cmap = 'viridis')
plt.title("Sepal Length vs Sepal Width (Iris Dataset)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.colorbar(label = "Species")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score



X = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
Y = df["Species"]



Y = Y.astype('category').cat.codes



X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)



model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X_train, Y_train)



predictions = model.predict(X_test)



accuracy = accuracy_score(Y_test, predictions)
print(f"Model Accuracy:{accuracy*100:.2f}%")
```

```python
with open("Iris_Analysis_Results.txt", "w") as file:
    file.write(f"Model Accuracy:{accuracy*100:.2f}%\n")
```
