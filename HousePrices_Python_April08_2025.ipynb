```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\House Prices.csv")




def avg_price_by_quality(data):
    return data.groupby('OverallQual')['SalePrice'].mean()




avg_prices = avg_price_by_quality(df)
print("Average Sale Price by Overall Quality :\n", avg_prices)
```

```python
class House:
    def __init__(self, sale_price, overall_qual, gr_liv_area, bedroom_abv_gr, full_bath, year_built):
        self.sale_price = sale_price
        self.overall_qual = overall_qual
        self.gr_liv_area = gr_liv_area
        self.bedroom_abv_gr = bedroom_abv_gr
        self.full_bath = full_bath
        self.year_built = year_built




    def display_info(self):
        print(f"House : Sale Price = {self.sale_price}, Overall Quality = {self.overall_qual}, "f"Living Area = {self.gr_liv_area} sqft, Bedrooms = {self.bedroom_abv_gr}, "f"Full Bath  {self.full_bath}, Year Built = {self.year_built}")




house1 = House(df['SalePrice'][0], df['OverallQual'][0], df['GrLivArea'][0], df['BedroomAbvGr'][0], df['FullBath'][0], df['YearBuilt'][0])
house1.display_info()
```

```python
plt.figure(figsize = (10, 6))
plt.plot(df['YearBuilt'], df['SalePrice'], marker = 'o', color = 'purple', label = 'Sale Price over Years')
plt.title('Sale Price of Houses by Year Built')
plt.xlabel('Year Built')
plt.ylabel('Sale Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```
