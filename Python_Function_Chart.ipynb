```python
X = [1,2,3,4,5]
Y = [10,20,25,30,35]



def calculate_total(X,Y):
    return X + Y



result = calculate_total(10,20)
print(f"Total:{result}")
```

```python
import matplotlib.pyplot as plt



X = [1,2,3,4,5]
Y = [10,20,25,30,35]



plt.plot(X,Y,label="Growth",color="green",marker="o")
plt.title("Simple Growth Chart")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
