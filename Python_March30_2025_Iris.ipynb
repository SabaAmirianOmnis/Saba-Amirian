```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\Iris.csv")




petal_dict = df.groupby('Species')['PetalLengthCm'].mean().to_dict()
print("Average Petal Length :", petal_dict)




plt.bar(petal_dict.keys(), petal_dict.values(), color = ['blue', 'green', 'purple'])
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```
