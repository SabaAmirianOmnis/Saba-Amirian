```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\StudentsPerformance.csv")




avg_scores = df.groupby("gender")[["math score", "reading score", "writing score"]].mean().to_dict()
print("Average Scores by Gender :", avg_scores)




plt.figure(figsize = (8, 6))
sns.heatmap(df[["math score", "reading score", "writing score"]].corr(), annot = True, cmap = "coolwarm")
plt.title("Correlation Heatmap of Scores")
plt.show()
```

```python
class Student:
    def __init__(self, name, math_score):
        self.name = name
        self.math_score = math_score




    def get_status(self):
        return "Pass" if self.math_score >= 50 else "Fail"




class AdvancedStudent(Student):
    def __init__(self, name, math_score, reading_score):
        super(). __init__(name, math_score)
        self.reading_score = reading_score




    def get_status(self):
        return "Pass" if self.math_score >= 50 and self.reading_score >= 50 else "Fail"




student1 = AdvancedStudent("Ali", 75, 80)
print(f"Student {student1.name} Status : {student1.get_status()}"}
```
