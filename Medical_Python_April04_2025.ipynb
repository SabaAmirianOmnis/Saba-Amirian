```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt




df = pd.read_csv(r"C:\Users\ASUS\Downloads\Medical.csv")




avg_charges = df.groupby("smoker")["charges"].mean().to_dict()
print("Average Charges by Smoker :", avg_charges)




plt.figure(figsize = (8, 6))
sns.heatmap(df[["age", "bmi", "charges"]].corr(), annot = True, cmap = "Blues")
plt.title("Correlation Heatmap of Medical Costs")
plt.show()
```

```python
class Patient:
    def __init__(self, age, charges):
        self.age = age
        self.charges = charges




    def get_risk(self):
        return "High" if self.charges > 20000 else "Low"




class DetailedPatient(Patient):
    def __init__(self, age, charges, smoker):
        super(). __init__(age, charges)
        self.smoker = smoker




    def get_risk(self):
        return "High" if self.charges > 20000 or self.smoker == "yes" else "Low"




patient1 = DetailedPatient(30, 25000, "yes")
```
